```python
import copy
```

# Modelagem

```python
xi = [i for i in range(0,5)]
fx= [i**4 for i in range(5)]
def Lx(indice, listax, X, xi):
    aux=1
    aux2=1

    for i in range(len(listax)):
        aux = aux * (X - listax[i])
        aux2 = aux2 * (xi[indice] - listax[i])
    return (aux/aux2)
    
def Polinomio_de_lagrange(xi,fxyz, X):
    auxx = 0
    for indice, f in enumerate(fxyz):
        listax = copy.deepcopy(xi)
        listax.remove(xi[indice])
        jj = f * Lx(indice, listax, X, xi)
        auxx += jj
    return (auxx)
```

# Dado certos pontos...

```python
import matplotlib.pyplot as plt


# Criando o gráfico
plt.scatter(xi, fx)

# Mostrando o gráfico
plt.show()
```

# O polinômio se ajusta com perfeição aos pontos

```python
# Criando o gráfico
xx = [i for i in range(0,5)]
plt.plot(xx, [Polinomio_de_lagrange(xi, fx, i) for i in xx])
plt.scatter(xx, [Polinomio_de_lagrange(xi, fx, i) for i in xx])

# Mostrando o gráfico
plt.show()
```

# Desvio 0 dos pontos

```python
# Criando o gráfico
xx = [i for i in range(-2,6)]
plt.plot(xx, [Polinomio_de_lagrange(xi, fx, i) for i in xx])
plt.scatter(xi,fx)

# Mostrando o gráfico
plt.show()
```
